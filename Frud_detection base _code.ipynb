#RandomForestClassifier

The Random Forest, also known as the Random Decision Forest, is a kind of supervised machine learning method that use decision trees to do classification, regression, as well as other types of tasks.
A random subset of the training data is used in the construction of the decision trees that are produced by the Random forest classifier. It is essentially a collection of decision trees (DT) that are drawn from a subset of the training set that is chosen at random. After that, it compiles the votes from the many decision trees in order to arrive at a conclusion on the final prediction.

```python
# we define classifier for random forest classifier with hyperparmeter.
rf = RandomForestClassifier( max_depth=5, random_state=0)
rf.fit(X_train, y_train)    # we fit classifier for random forest
y_pred = rf.predict(X_test) # we predict classifier for random forewst.
```

```python
print('Accuracy: %.3f' % (accuracy_score(y_test,y_pred)*100))
###################################################################

"""The F1 score for a given set of predicted labels may be computed in Python using the f1 score function that is included in the sklearn metrics module.
 The possible values for an F1 score are 0 to 1, with 0 being the poorest possible score and 1 representing the best possible score."""
print('F1_Score: %.3f' % (f1_score(y_test,y_pred, average='weighted')*100)) 

##############################################
"""The term "accuracy of the test" refers to the ratio of the number of real positives to the overall number of positives,
 which includes the number of false positives. Intuitively, accuracy is the ability of the classifier to avoid incorrectly labelling as positives an example that actually has a value that is negative."""

print('Precision_Score : %.3f' % (precision_score(y_test, y_pred,average='weighted')*100))

#######################################################################################
print('Recall_score : %.3f' % (recall_score(y_test, y_pred,average='weighted')*100))
```

#Sensitivity
The degree to which a machine learning algorithm is able to recognise good examples is referred to as its sensitivity. In certain circles, it is referred to as the "true positive rate," or recall. When evaluating the performance of a model, sensitivity is one of the metrics that is considered since it shows how many positive examples the model had able to accurately detect.

#Specificity 
Both sensitivity and specificity are two measures that are used in the area of machine learning to assess the efficacy of a model. The model's sensitivity refers to the proportion of real positives that can be accurately predicted by the model, while the model's specificity refers to the proportion of genuine negatives that can be accurately predicted by the model.

```python

cm1 = confusion_matrix(y_test, y_pred)
# printing Sensitivity accuracy 
sensitivity = cm1[0,0]/(cm1[0,0]+cm1[0,1])
print('Sensitivity : ', sensitivity)
# printing Specificity accuracy 
specificity = cm1[1,1]/(cm1[0,1]+cm1[1,1])
print('Specificity : ', specificity)
```

A traditional neural network layer that has high coupling is referred to as a dense layer. It is the layer that is discovered and used the most. The activity described in the next paragraph is carried out on the input by the dense layer, which then sends back the result.

The neural network is responsible for the activation of functions. The activation function has an effect on the firing of neurons, as the graphic illustrates.

```python
input_dim = X_train1.shape[1]
encoding_dim = 80
input_layer = Input(shape=(input_dim, ))
encoder_model = Dense(encoding_dim*2, activation="sigmoid")(input_layer)
#The neural network is responsible for the activation of functions. The activation function has an effect on the firing of neurons, as the graphic illustrates.
encoder_model = Dense(encoding_dim, activation="sigmoid")(input_layer)
#A traditional neural network layer that has high coupling is referred to as a dense layer. It is the layer that is discovered and used the most. The activity described in the next paragraph is carried out on the input by the dense layer, which then sends back the result.
encoder_model = Dense(1020,activation="sigmoid")(encoder_model)
decoder_model = Dense(565, activation='sigmoid')(encoder_model)
decoder_model = Dense(1020, activation='sigmoid')(encoder_model)
decoder_model = Dense(input_dim, activation='sigmoid')(decoder_model)
autoencoder_model = Model(inputs=input_layer, outputs=decoder_model)
```

The Adaptive Movement Estimation method, sometimes known as Adam for short, is an extension to the gradient descent algorithm and a logical successor to approaches such as AdaGrad and RMSProp. Adam is also known as the method for estimating adaptive movement. This method will first automatically modify a learning rate for each and every input variable that is used to calculate the objective function. After that, it will further smooth out the search process by using an exponentially smooth search.

```python
nb_epoch = 40
batch_size =250
import tensorflow as tf
adam = tf.keras.optimizers.Adam(learning_rate=0.001)
autoencoder_model.compile(optimizer=adam, loss='mean_squared_error', metrics=['accuracy'])
history1 = autoencoder_model.fit(X_train1, X_train1,epochs=nb_epoch, batch_size=batch_size,validation_data=(X_test1, X_test1), verbose=1)
```

```python
#plotting  loss figure 
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
plt.title('model loss')# set title for graph.
plt.ylabel('loss')# set y label for graph.
plt.xlabel('epoch')# set x label for graph.
plt.legend(['train', 'test'], loc='upper right')
plt.show() # it is used to show the graph
# plotting accuracy figure
plt.plot(history1.history['accuracy'])
plt.plot(history1.history['val_accuracy'])
plt.title('model acc')   # set title for graph.
plt.ylabel('accuracy')  # set y label for graph.
plt.xlabel('epoch')   # set x label for graph.
plt.legend(['train', 'test'], loc='upper right')
plt.show()  # it is used to show the graph.
```

```python
predictions = autoencoder_model.predict(X_test1)
mse = np.mean(np.power(X_test1 - predictions, 2), axis=1)
error_df = pd.DataFrame({'reconstruction_error': mse,
                        'true_class': y_test1})
error_df.describe()
```
